Q01

```python
age=float(input("What is your age? (In Years):"))

if age<1: print("He/she is baby.")
elif age<3: print("He/she is toddler.")
elif age<14: print("He/she is kid.")
elif age<20: print("He/she is teen.")
elif age<60: print("He/she is adult.")
else: print("He/she is elder.")
```

```python
year=int(input("Enter the age:"))
f1=year%100
f11=year%400
f2=year%4
if f1==0:
  if f11==0: print(year, " is a leap year")
  else: print(year, " is not a leap year")
else:
  if f2==0: print(year, " is not a leap year.")
  else: print(year, " is not a leap year")
```
