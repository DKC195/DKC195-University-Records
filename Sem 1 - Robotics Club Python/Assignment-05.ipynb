Q1

```python
x=int(input("Enter the no. of input: "))
num=[]
for i in range(x):
  num.append(int(input("Enter a number: ")))
even=list(filter(lambda x: x%2!=0,num))
odd=list(filter(lambda x: x%2==0,num))

print("The even numbers are: ",even)
print("The odd numbers are: ",odd)

```

```python
x=int(input("Enter the no. of input: "))
num=[]
for i in range(x):
  num.append(int(input("Enter a number: ")))
s=list(map(lambda a: a*a,num))
c=list(map(lambda a: a*a*a,num))
print("The square are", end=" ")
print(s, end=" ")
print("and the cube are", end=" ") 
print(c)
```

```python
x=int(input("Enter the no. of words: "))
words=[]
for i in range(x):
  words.append(input("Enter a number: "))
y=input("Enter a word to remove: ")
out=list(filter(lambda words: words not in y, words ))
print(out)
```
