Q03

```python
menu={"nori": 100, "bread":90, "milk":90, "chicken":300, "mutton":1000, "beef":3000, "chocolate":100, "oil":120, "coke":150}
print("Menu")
for key,price in menu.items():
  key=key.capitalize()
  print(key, ':' , price)

print("")
bill={}

inp="y"
add=0
while (inp.lower()=="y"):
  order=input("What did you buy?:")
  unit=menu[order.lower()]
  quantity=int(input("How many did you buy?:"))
  inp=input("Do you still want to keep ordering? (Y/N):")
  cost=unit*quantity
  bill[order.capitalize()]=cost
  add=add+cost

print("")
print("---Your Bill---")
for key,price in bill.items():
  key=key.capitalize()1
  print(key, ':' , price)
print("------------------------------")
print("Total",':',add)
```
